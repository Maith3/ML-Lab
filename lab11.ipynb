```python
import numpy as np
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
class Perceptron:
  def __init__(self,input_size):
    self.weights = np.random.rand(input_size)
    self.bias = np.random.rand(1)
  def forward(self,input):
    total_input = np.dot(self.weights,input)+self.bias
    return sigmoid(total_input)
  def train(self,X,y,epochs=1000,learning_rate=0.1):
    for epoch in range(epochs):
      for i in range(X.shape[0]):
        output = self.forward(X[i])
        error = y[i]-output
        self.weights += learning_rate*error*X[i]
        self.bias += learning_rate*error
```

```python
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y_and = np.array([0,0,0,1])
y_or = np.array([1,1,1,0])
```

```python
perceptron_or = Perceptron(input_size=2)
perceptron_and = Perceptron(input_size=2)
```

```python
perceptron_or.train(X,y_or)
perceptron_and.train(X,y_and)
```

```python
for i in range(X.shape[0]):
  print(np.round(perceptron_and.forward(X[i])))
```

```python
for i in range(X.shape[0]):
  print(np.round(perceptron_or.forward(X[i])))
```
